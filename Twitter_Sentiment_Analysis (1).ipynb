```python
#dropping the columns which have numeric data
text_df = df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'], axis=1)
text_df.head()
```

```python
stop_words = set(stopwords.words('english'))
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head()
```

```python
def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

# Create a CountVectorizer object with n-gram range (1, 2)
vect = CountVectorizer(ngram_range=(1, 2))

# Fit the CountVectorizer to the text data
vect.fit(text_df['text'])

# Access feature names using get_feature_names() method
feature_names = vect.get_feature_names_out()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n{}".format(feature_names[:20]))

X = text_df['text']
Y = text_df['sentiment']

# Transform the text data using the fitted CountVectorizer
X = vect.transform(X)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print('Size of x_tarin:',(x_train.shape))
print('Size of y_train:',(y_train.shape))
print('Size of x_test:',(x_test.shape))
print('Size of y_test:',(y_test.shape))
```

```python
lr = LogisticRegression() #lr stands for logistic regression
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_accuracy = accuracy_score(lr_pred, y_test)
print("Accuracy: {:.2f}%".format(lr_accuracy * 100))
```

```python
print(confusion_matrix(y_test, lr_pred))
print("\n")
print(classification_report(y_test, lr_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(y_test, lr_pred, labels=lr.classes_)#cm stands for confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=lr.classes_)
disp.plot()
```
